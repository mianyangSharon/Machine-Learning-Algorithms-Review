## Cosine Similarity

In Machine Learning, features are often in vector form. 

Cosine similarity is often used when analyzing the similarity between two feature vectors.

Cosine similarity is between [-1, 1]

Cosine similarity between two same vectors is 1.


$$ cos(A, B) = \frac{A*B}{||A||_2||B||_2} $$

* Cosine similarity only cares the angel between two vectors, the size of the vectors do not matter

* Q: When to use Cosine Similarity not Euclidean Distance?
* A: In text, picture, video analysis, the feature dimensions are usually very high. Cosine similarity in high dimensions is still [-1,1], while Euclidean Distance varies.


```python
# A, B are 1-d array of same size
def cosine_similarity(A, B):
    if len(A) < 2 or len(B) < 2:
        print("invalid vectors")
        return 2
    return np.dot(A,B)/(np.linalg.norm(A)*np.linalg.norm(B))

print (cosine_similarity(np.asarray([1,2,3]), np.asarray([2,3,4])))
```

## Cosine Distance

$$ Cosine Distance = 1 - Cosine Similarity $$

Thus cosine distance is between [0, 2]

Cosine distance between two same vectors is 0.

* Q: When to use Cosine Distance, when to use Euclidean Distance?
* A: 
* Cosine Distance - the relative distance between two directions
* Euclidean Distance - the absolute distance between two values
    * E.g. if to analyze the behavior of watching movies between two costumers - use Cosine Distance
    * E.g. if to analyze the Activation of users, using (times logged in, average watching time length) - use Euclidean
    
  For example if we get (1,10) and (10, 100), Cosine Distance does not make sense here... they are same direction but very different behaviors!


```python
# A, B are 1-d array of same size
def cosine_distance(A, B):
    return 1 - cosine_similarity(A, B)
print (cosine_distance(np.asarray([1,2,3]), np.asarray([2,3,4])))
```

## Euclidean Distance

```python
## Euclidean Distance
def euclidean_distance(A, B):
    # or using np.linalg.norm(A, B)
    return np.sqrt(np.sum((A - B)**2))
```

```python
print (euclidean_distance(np.asarray([1,2,3]), np.asarray([2,3,4])))
```
